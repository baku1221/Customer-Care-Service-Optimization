Calculating and Plotting the number of calls during :

*   Different Days of the Week
*   Different Days of the Month
*   Different Months of the Year
*   Different Hours of the Day



```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming the data is loaded into a dataframe named 'df'
# Load your dataframe


# Convert the date columns to datetime format
calls['call_start_datetime'] = pd.to_datetime(calls['call_start_datetime'])
calls['call_end_datetime'] = pd.to_datetime(calls['call_end_datetime'])

# 1. Number of calls in different hours of the day
calls['call_hour'] = calls['call_start_datetime'].dt.hour
hourly_calls = calls.groupby('call_hour').size()

# Plotting the number of calls per hour
plt.figure(figsize=(10, 5))
hourly_calls.plot(kind='bar', color='skyblue')
plt.title('Number of Calls per Hour of the Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Calls')
plt.xticks(rotation=0)
plt.show()

# 2. Number of calls in different months of the year
calls['call_month'] = calls['call_start_datetime'].dt.month
monthly_calls = calls.groupby('call_month').size()

# Plotting the number of calls per month
plt.figure(figsize=(10, 5))
monthly_calls.plot(kind='bar', color='lightgreen')
plt.title('Number of Calls per Month')
plt.xlabel('Month')
plt.ylabel('Number of Calls')
plt.xticks(rotation=0)
plt.show()

# 3. Number of calls on different days of the week
calls['call_day_of_week'] = calls['call_start_datetime'].dt.day_name()
weekly_calls = calls.groupby('call_day_of_week').size()

# Plotting the number of calls per day of the week
plt.figure(figsize=(10, 5))
weekly_calls.plot(kind='bar', color='lightcoral')
plt.title('Number of Calls per Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Calls')
plt.xticks(rotation=45)
plt.show()

calls['call_day_of_month'] = calls['call_start_datetime'].dt.day
daily_calls = calls.groupby('call_day_of_month').size()

# Plotting the number of calls per day of the month
plt.figure(figsize=(10, 5))
daily_calls.plot(kind='bar', color='orange')
plt.title('Number of Calls per Day of the Month')
plt.xlabel('Day of the Month')
plt.ylabel('Number of Calls')
plt.xticks(rotation=0)
plt.show()
```

Calculating the Call Duration in Minutes - using Call Start Time and Call End Time

```python
calls['call_start_datetime'] = pd.to_datetime(calls['call_start_datetime'])
calls['call_end_datetime'] = pd.to_datetime(calls['call_end_datetime'])

# Calculate the call duration in seconds
calls['call_duration'] = calls['call_end_datetime'] - calls['call_start_datetime']

# To get the duration in minutes or seconds
calls['call_duration_minutes'] = calls['call_duration'].dt.total_seconds() / 60

# Print the resulting DataFrame
print(calls[['call_start_datetime', 'call_end_datetime', 'call_duration_minutes']])
```

Sentiment Analysis and Call Tone - Customer and Agent Tone.

```python
sentiment_stats = pd.read_csv('/content/drive/MyDrive/UnitedData/sentiment_statisticscc1e57a.csv')
```

```python
sentiment_stats
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame structure (replace this with your actual dataframe)
# sentiment_stats = pd.DataFrame({
#     'call_id': [...],
#     'agent_id': [...],
#     'customer_tone': [...],
#     'agent_tone': [...],
#     'call_silence_time': [...]
# })

### 1. Distribution of Customer and Agent Tones

# Count the occurrences of each tone for customers and agents
customer_tone_counts = sentiment_stats['customer_tone'].value_counts()
agent_tone_counts = sentiment_stats['agent_tone'].value_counts()

# Plot the distribution of customer tones
plt.figure(figsize=(8, 4))
sns.barplot(x=customer_tone_counts.index, y=customer_tone_counts.values, palette='coolwarm')
plt.title('Distribution of Customer Tones')
plt.xlabel('Customer Tone')
plt.ylabel('Count')
plt.show()

```

```python

# Plotting the pie chart for customer tone distribution
plt.figure(figsize=(6, 6))
plt.pie(customer_tone_counts, labels=customer_tone_counts.index, autopct='%1.1f%%', colors=['red', 'orange', 'yellow', 'green', 'blue'], startangle=140, shadow=True)

# Add a title
plt.title('Customer Tone Distribution')

# Show the plot
plt.show()
```

```python
# Plotting the pie chart for customer tone distribution
plt.figure(figsize=(7, 7))
plt.pie(agent_tone_counts, labels=agent_tone_counts.index, autopct='%1.1f%%', colors=['red', 'orange', 'yellow', 'green', 'blue'], startangle=140, shadow=True, textprops={'fontsize': 7})

# Add a title
plt.title('Agent Tone Distribution')

# Show the plot
plt.show()
```

```python
# Plot the distribution of agent tones
plt.figure(figsize=(8, 4))
sns.barplot(x=agent_tone_counts.index, y=agent_tone_counts.values, palette='coolwarm')
plt.title('Distribution of Agent Tones')
plt.xlabel('Agent Tone')
plt.ylabel('Count')
plt.show()
```

```python
### 2. Correlation Between Customer and Agent Tones

# Create a crosstab to show the relationship between customer tone and agent tone
tone_correlation = pd.crosstab(sentiment_stats['customer_tone'], sentiment_stats['agent_tone'])

# Plot the heatmap for tone correlation
plt.figure(figsize=(6, 4))
sns.heatmap(tone_correlation, annot=True, cmap='Blues', fmt='d')
plt.title('Correlation between Customer and Agent Tones')
plt.xlabel('Agent Tone')
plt.ylabel('Customer Tone')
plt.show()
```

```python
### 3. Call Silence Time Distribution

# Plot the distribution of call silence time
plt.figure(figsize=(8, 4))
sns.histplot(sentiment_stats['silence_percent_average'], bins=20, kde=True, color='purple')
plt.title('Distribution of Call Silence Time (as %)')
plt.xlabel('Silence Time (%)')
plt.ylabel('Number of Calls')
plt.show()
```

```python
### 4. Silence Time vs. Tone

# Boxplot to compare silence time with customer tone
plt.figure(figsize=(8, 4))
sns.boxplot(x='customer_tone', y='silence_percent_average', data=sentiment_stats, palette='Set2')
plt.title('Call Silence Time by Customer Tone')
plt.xlabel('Customer Tone')
plt.ylabel('Silence Time (%)')
plt.show()
```

```python
# Boxplot to compare silence time with agent tone
plt.figure(figsize=(8, 4))
sns.boxplot(x='agent_tone', y='silence_percent_average', data=sentiment_stats, palette='Set2')
plt.title('Call Silence Time by Agent Tone')
plt.xlabel('Agent Tone')
plt.ylabel('Silence Time (%)')
plt.show()
```

```python
### 5. Agent Performance (Average Silence Time and Tones)

# Group by agent_id to get the average silence time for each agent
agent_silence_avg = sentiment_stats.groupby('agent_id')['silence_percent_average'].mean().reset_index()

# Count positive, neutral, and negative tones for each agent
agent_tone_summary = sentiment_stats.groupby('agent_id')['agent_tone'].value_counts(normalize=True).unstack().fillna(0)

# Merge both into a single DataFrame for comparison
agent_performance = agent_silence_avg.merge(agent_tone_summary, on='agent_id')

# Plot agent performance (silence time vs. tone proportions)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='silence_percent_average', y='calm', data=agent_performance, s=10, color='green')
sns.scatterplot(x='silence_percent_average', y='frustrated', data=agent_performance, s=10, color='red')
sns.scatterplot(x='silence_percent_average', y='neutral', data=agent_performance, s=10, color='yellow')
sns.scatterplot(x='silence_percent_average', y='polite', data=agent_performance, s=10, color='purple')
sns.scatterplot(x='silence_percent_average', y='angry', data=agent_performance, s=10, color='orange')



plt.title('Agent Performance: Call Silence Time vs. Tone Proportions')
plt.xlabel('Average Silence Time (%)')
plt.ylabel('Tone Proportion')
plt.legend(['Calm', 'Frustrated',"Neutral","Polite","Angry"])
plt.show()
```

```python
import pandas as pd

# Example: Assuming customer_tone has these possible values: 'calm', 'angry', 'frustrated', 'polite', 'neutral'

# Group by 'agent_id' and 'customer_tone', then count occurrences
tone_counts = sentiment_stats.groupby(['agent_id', 'customer_tone']).size().unstack(fill_value=0)

# Ensure all required tone columns exist, even if empty for some agents
required_tones = ['calm', 'angry', 'frustrated', 'polite', 'neutral']
for tone in required_tones:
    if tone not in tone_counts.columns:
        tone_counts[tone] = 0

# Create the final DataFrame, filling missing values with 0
tone_counts = tone_counts[required_tones].reset_index()

# Rename columns for clarity
tone_counts.columns = ['agent_id', 'calm_tone', 'angry_tone', 'frustrated_tone', 'polite_tone', 'neutral_tone']

# Print the new DataFrame
tone_counts
```
